## Preprocessing Design

### Image Size
All MRI images will be resized to 224×224 pixels.  
This size is chosen because it is standard for many pretrained CNN architectures and balances detail with computational efficiency.

### Normalization
Images will be normalized using ImageNet-style normalization (mean and standard deviation).  
This allows better compatibility with pretrained models and stabilizes training.

### Data Augmentation
To improve generalization and reduce overfitting, the following augmentations will be applied to training data:
- Horizontal flip
- Small random rotations

### Dataset Splitting
The dataset will be split into:
- Training set
- Validation set
- Test set

Where possible, stratified splitting will be used to preserve class distribution across splits.
